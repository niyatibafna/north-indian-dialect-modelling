```python
import texttable as tt
import latextable as lt
import json
import pandas as pd
from collections import defaultdict
```

```python
with open("../evaluation_languages_home/eval_results/mli_ned/results_hindi-urdu_source_softgold.json", "r") as f:
    softgold = json.load(f)
with open("../evaluation_languages_home/eval_results/mli_ned/results_hindi-urdu_source_loose.json", "r") as f:
    loose = json.load(f)
with open("../evaluation_languages_home/eval_results/mli_jw/results_hindi-urdu_source_softgold.json", "r") as f:
    jw_softgold = json.load(f)
with open("../evaluation_languages_home/eval_results/mli_jw/results_hindi-urdu_source_loose.json", "r") as f:
    jw_loose = json.load(f)
with open("../evaluation_languages_home/eval_results/mli_sem_od/results_hindi-urdu_source_loose.json", "r") as f:
    so_loose = json.load(f)
with open("../evaluation_languages_home/eval_results/mli_sem_od/results_hindi-urdu_source_softgold.json", "r") as f:
    so_softgold = json.load(f)
```

```python
results = defaultdict(lambda: dict())
for lang in softgold:
    if lang == "hindi-urdu":
        continue
    results[lang]["found"] = softgold[lang]["found"]
    results[lang]["total"] = softgold[lang]["total"]
#     results[lang]["NED Softgold acc."] = round(softgold[lang]["accuracy"], 2)
    results[lang]["NED Loose acc."] = round(loose[lang]["accuracy"], 2)
#     results[lang]["JW Softgold acc."] = round(jw_softgold[lang]["accuracy"], 2)
    results[lang]["JW Loose acc."] = round(jw_loose[lang]["accuracy"], 2)
    results[lang]["Sem OD Loose acc."] = round(so_loose[lang]["accuracy"], 2)
#     results[lang]["Sem OD Softgold acc."] = round(so_softgold[lang]["accuracy"], 2)
    
```

```python

with open("results_mli_ned,JW,SemOD_loose.json", "w") as f:
    json.dump(results, f, indent=2)
```

```python
df = pd.DataFrame(results).transpose()
# df = df.values.tolist()
df
# t1 = tt.Texttable()
# t1.add_rows(df)
# t1.draw()
# print(lt.draw_latex(t1))
```
