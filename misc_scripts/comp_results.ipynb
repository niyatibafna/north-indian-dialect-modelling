```python
import pandas as pd
import json
from collections import defaultdict, Counter
```

```python
res_file1 = "../evaluation_languages_home/eval_results/mli_ned/results_top5_hindi-urdu_source_loose.json"
res_file2 = "../evaluation_languages_home/eval_results/mli_jw/results_top5_hindi-urdu_source_loose.json"
res_file3 = "../evaluation_languages_home/eval_results/mli_em_od/results_top5_hindi-urdu_source_loose.json"
res_file4 = "../evaluation_languages_home/eval_results/mli_sem_od/results_top5_hindi-urdu_source_loose.json"
res_file5 = "../evaluation_languages_home/eval_results/mli_sem_emod/results_top5_target0_hindi-urdu_source_loose.json"

lab_1 = "NED"
lab_2 = "JWM"
lab_3 = "em_od"
lab_4 = "sem_od"
lab_5 = "sem_emod"
```

```python

with open(res_file1, "r") as f:
    res1 = json.load(f)

with open(res_file2, "r") as f:
    res2 = json.load(f)

with open(res_file3, "r") as f:
    res3 = json.load(f)
    
with open(res_file4, "r") as f:
    res4 = json.load(f)

with open(res_file5, "r") as f:
    res5 = json.load(f)
```

```python
r = 3
results = defaultdict(lambda: dict())
for lang in res1:
    if lang == "hindi-urdu":
        continue
    results[lang]["Total"] = round(res1[lang]["total"], 0)
    results[lang]["Fotal"] = round(res1[lang]["found"], 0)
    results[lang][lab_1] = round(res1[lang]["accuracy"]*100, r)
    results[lang][lab_2] = round(res2[lang]["accuracy"]*100, r) if lang in res2 else "-"
    results[lang][lab_3] = round(res3[lang]["accuracy"]*100, r) if lang in res3 else "-"
    results[lang][lab_4] = round(res4[lang]["accuracy"]*100, r) if lang in res4 else "-"
    results[lang][lab_5] = round(res5[lang]["accuracy"]*100, r) if lang in res5 else "-"
```

```python
avg = dict()
for l in [lab_1, lab_2, lab_3, lab_4, lab_5]:
    avg[l] = sum([results[lang][l] for lang in results])/len(results)
avg["Total"] = sum([results[lang]["Total"] for lang in results])/len(results)
avg["Found"] = sum([results[lang]["Fotal"] for lang in results])/len(results)
    
```

```python
avg_all = {"Avg.": avg}
```

```python
df = pd.DataFrame(avg_all).transpose()
print(df.to_latex(caption = "Average results over all languages per approach", label ="tab:res_avg"))
```
