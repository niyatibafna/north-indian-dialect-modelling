```python
import fasttext
import sys
import math
```

```python
langs = [

# "rajasthani",
"hariyanvi",
"punjabi",
"hariyanvi",
# "gujarati",
"khadi_boli",
"sanskrit",
# "hindi-urdu",
"bhil",
"koraku",
"baiga",
"nimaadi",
"malwi",
"marathi",
"bhadavari",
"himachali",
"garwali",
"kumaoni",
"kannauji",
"brajbhasha",
"bundeli",
"awadhi",
"chattisgarhi",
"nepali",
"pali",
"bhojpuri",
"bajjika",
"magahi",
"maithili",
"angika",
"sindhi",
# "bangla"
        ]
```

```python
DATAPATH = "../../data/crawled_cleaned/"
MODEL_DIR = "../models/upsampled/"
TEMP_DIR = "../../data/temp_upsampled/"
OUT_DIR = "../../data/upsampled/"
anch = "hindi-urdu"
threshold = 3

for lang in langs:

    LANG_PATH = DATAPATH + lang + ".txt"
    ANCH_PATH = DATAPATH + anch + ".txt"
    MODEL_PATH = MODEL_DIR + "{}_{}.bin".format(lang, anch)
    TEMP_PATH = TEMP_DIR + "{}_{}.txt".format(lang, anch)
    OUT_PATH = OUT_DIR + "{}.txt".format(lang)
    


    lang_data = open(LANG_PATH, "r").read()
    anch_data = open(ANCH_PATH, "r").read()

    len_ldata = int(math.log(len(lang_data.split()), 10))
    len_adata = int(math.log(len(anch_data.split()), 10))


    ratio = 10**(len_adata - len_ldata)

    up_lang_data = " ".join(lang_data.split()*ratio)

    with open(OUT_PATH, "w") as f:
        f.write(up_lang_data)

    print("LENGTH OF {}: {}".format(anch , len(anch_data.split())))
    print("LENGTH OF {}: {}".format(lang , len(lang_data.split())))
    print("RATIO : {}".format(ratio))
    print("UPSAMPLED LEN OF {}: {}".format(lang, len(up_lang_data.split())))

    all_data = anch_data + " " + up_lang_data
    with open(TEMP_PATH, "w") as f:
        f.write(all_data)


    model = fasttext.train_unsupervised(TEMP_PATH, dim = 150, minCount = 3)
    model.save_model(MODEL_PATH)
```
