```python
import json
```

```python
with open("hindi2nepali_bil_lex.json", "r") as f:
    hin2nep_bl = json.load(f)
```

```python
bornates = dict()
for source, target in hin2nep_bl.items():
    print("HINDI: ", source)
    print("NEPALI: ", target)
    verdict = int(input())
    if verdict == 1:
        bornates[source] = target
    if verdict == 0:
        break
    if len(bornates) == 100:
        break
```

with open("hindi2nepali_bornates.json", "w") as f:
    json.dump(bornates, ensure_ascii=False, indent = 2)
