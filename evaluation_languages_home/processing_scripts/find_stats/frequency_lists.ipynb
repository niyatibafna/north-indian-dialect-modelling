```python
import os
import string
from collections import Counter
import json
```

```python
DIR = "../../eval_data/formatted_eng/"
```

```python
def clean(text):
    eng_words = ["respect", "he", "she", "plural", "to",\
    "sit", "eat", "walk", "drink", "go", "m", "f"]
    text = " ".join(text)
    text = "".join([c if c not in string.punctuation else " " for c in text])
    text = [word for word in text.split(" ") if word not in eng_words]
    return text
```

```python
lexicons = dict()
for file in os.listdir(DIR):
    lang = file.split(".")[0]
    par_sent = open(DIR+file, "r").read().split("\n")
    target = [sent.split("|||")[0].strip() for sent in par_sent]
    target = clean(target)
    lexicons[lang] = Counter(target)
    
```

```python
for lang in lexicons:
    lexicons[lang] = sorted(lexicons[lang].items(), key = lambda x: x[1], reverse = True)
```

```python
with open("../../eval_data/stats/word_frequencies_againsteng.json", "w") as f:
    json.dump(lexicons, f)
```
