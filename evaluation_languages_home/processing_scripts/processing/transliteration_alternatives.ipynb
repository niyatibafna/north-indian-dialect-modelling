```python
from indictrans import Transliterator
import os
import json
from collections import Counter, defaultdict
import pandas as pd
import editdistance
```

```python
trn = Transliterator(source = "eng", target = "hin", build_lookup = True)
```

```python
vowel_range = list(range(2305, 2315)) + list(range(2317, 2325)) + list(range(2365, 2384))
bad_char_range = range(2364, 2367)

def find_closest(word, lang, lexicon_corpus):
    '''Find closest word in corpus of lang'''
    min_dist, best_word = 2, ""
    for cand in lexicon_corpus[lang]:
        if cand[0] != word[0]:
            continue
        cons_sequence_1 = "".join([c for c in word if ord(c) not in bad_char_range])
        cons_sequence_2 = "".join([c for c in cand if ord(c) not in bad_char_range])
        if cons_sequence_1 != cons_sequence_2:
            continue
        
        if editdistance.eval(word, cand)/max(len(word), len(cand)) < min_dist:
            min_dist = editdistance.eval(word, cand)/max(len(word), len(cand))
            best_word = cand
            
    return best_word, min_dist
```

```python
# Reading corpora 
DATA_DIR = "../../../data/crawled_cleaned/"
lexicon_corpus = dict()

for file in os.listdir(DATA_DIR):
    lang = file.split(".")[0]
    lexicon_corpus[lang] = Counter(open(DATA_DIR+file, "r").read().split(" "))
```

```python
with open("../../eval_data/stats/word_frequencies_againsteng.json", "r") as f:
    lexicon_test = json.load(f)
```

```python
lexicon_test = {lang: set([word[0] for word in lexicon_test[lang]]) for lang in lexicon_test}
```

```python
for lang in lexicon_corpus:
    if lang not in lexicon_test:
        continue
    print("NOW DOING: ", lang)
    non_zero = 0
    for word in lexicon_test[lang]:
        if word == "":
            continue
        trans_word = trn.transform(word)
        best_word, min_dist = find_closest(trans_word, lang, lexicon_corpus)
        if min_dist == 0 or min_dist > 1:
            continue
        non_zero += 1
        print("OG ", word)
        print("Trans word ", trans_word)
        print("Best word ", best_word)
        print("Min dist ", min_dist)
    
    print("Alternatives found: ", non_zero)
    print("\n\n\n")
```
